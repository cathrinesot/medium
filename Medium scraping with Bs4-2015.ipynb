```python
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import os
import requests
from datetime import datetime
import ast
```

```python
li = ['https://entrepreneurs.maqtoob.com/the-37-best-websites-to-learn-something-new-895e2cb0cad4',
'https://byrslf.co/when-i-m-gone-f1611ceb759f',
'https://medium.com/found-it/advice-from-30-year-old-me-to-20-year-old-me-b9b035d39e2d',
'https://m.signalvnoise.com/reconsider-41adf356857f',
'https://medium.com/@elleluna/the-crossroads-of-should-and-must-90c75eb7c5b0',
'https://medium.com/@bchesky/7-rejections-7d894cbaa084',
'https://entrepreneurshandbook.co/i-sat-down-with-a-millionaire-who-operates-10-businesses-while-sailing-around-the-world-with-his-338929c4e8c9',
'https://medium.com/bad-words/why-twitter-s-dying-and-what-you-can-learn-from-it-9ed233e37974',
'https://medium.com/the-global-future-of-work/there-is-something-extraordinary-happening-10492495c715',
'https://medium.com/@awilkinson/slack-s-2-8-billion-dollar-secret-sauce-5c5ec7117908',
'https://medium.com/swlh/300-awesome-free-things-e07b3cd5fd5b',
'https://m.signalvnoise.com/the-day-i-became-a-millionaire-55d7dc4d8293',
'https://blog.medium.com/taking-medium-to-the-next-level-cb7f223fad86',
'https://medium.com/@wilw/7-things-i-did-to-reboot-my-life-a4bab2d409e',
'https://medium.com/@sundar_pichai/let-s-not-let-fear-defeat-our-values-af2e5ca92371',
'https://medium.com/@jcheiffetz/i-had-a-baby-and-cancer-when-i-worked-at-amazon-this-is-my-story-9eba5eef2976',
'https://medium.com/backchannel/a-teenagers-view-on-social-media-1df945c09ac6',
'https://medium.com/@erikdkennedy/7-rules-for-creating-gorgeous-ui-part-1-559d4e805cda',
'https://medium.com/@kollinz/misused-mobile-ux-patterns-84d2b6930570',
'https://medium.com/facebook-design/how-we-changed-the-facebook-friends-icon-dc8526ea9ea8']
```

```python
list=[]
for i in li:
    dic = {}
    print(i)
    r=requests.get(i)
    data=r.text
    soup=BeautifulSoup(data,'lxml')
    
    
    dic['title'] = soup.title.text[0:-23]
    dic['claps'] = soup.find('button', class_="button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton").text
    format = "%Y-%m-%d"
    dic['datePublished'] = datetime.strptime(ast.literal_eval(soup.find('script').text)["datePublished"][0:10], format)
    dic['read_time'] = soup.find("span",class_ = "readingTime")['title'][:-9]
    dic['followedBy'] = soup.find_all("script")[-1].text.split('"usersFollowedByCount":')[1].split(",")[0]
    dic['following'] = soup.find_all("script")[-1].text.split('"usersFollowedCount":')[1].split(",")[0]
    #intro = soup.find('div', attrs={'class': 'u-lineHeightTightest'})
    
    dic['author'] = soup.find('a', attrs={'class': 'ds-link ds-link--styleSubtle ui-captionStrong u-inlineBlock link link--darken link--darker'}).text
    dic['body'] = [body.text.strip() for body in soup.find_all('p')]
    dic['tags'] = ast.literal_eval(soup.find('script').text)["keywords"]
    
    list.append(dic)
    print(dic)
    
    time.sleep(20)
    
```

```python
import pandas as pd
df = pd.DataFrame(list)
df
```

```python
df.to_csv("scrapped_2015.csv", index = False)
```
